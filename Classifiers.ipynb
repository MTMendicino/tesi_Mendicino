```python
import numpy as np 
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap #Questo pu essere molto utile quando 
#si indicizza direttamente in una mappa di colori, ma pu anche 
#essere usato per generare mappe di colori speciali per la mappatura ordinaria.

from sklearn.model_selection import train_test_split #divide i dati tra un gruppo di train e un gruppo di test
from sklearn.preprocessing import StandardScaler #sottrai la media e dividi per la varianza dei dati 
from sklearn.datasets import make_moons, make_circles, make_classification # non ho ben capito??????
from sklearn.neural_network import MLPClassifier #MLPClassifier si addestra in modo 
# iterativo poich ad ogni passo di tempo vengono calcolate le derivate parziali della 
# funzione di perdita rispetto ai parametri del modello per aggiornare i parametri.

from sklearn.neighbors import KNeighborsClassifier #Il k-nearest neighbors (k-NN)  un algoritmo utilizzato 
# nel riconoscimento di pattern per la classificazione di oggetti basandosi sulle caratteristiche degli
# oggetti vicini a quello considerato.

from sklearn.svm import SVC #un algoritmo di addestramento per le SVM costruisce un modello che
#assegna i nuovi esempi a una delle due classi, ottenendo quindi un classificatore lineare binario non probabilistico.

from sklearn.gaussian_process import GaussianProcessClassifier #metodo di apprendimento supervisionato generico 
#progettato per risolvere problemi di regressione e classificazione probabilistica .

from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier#Un classificatore ad albero decisione
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier #da utilizzare con i diagrammi ad albero
from sklearn.naive_bayes import GaussianNB 
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis


#librerie giulia 
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_curve, auc
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping
import tensorflow as tf
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
from hyperopt import fmin, tpe, hp, Trials
from hyperopt import STATUS_OK, space_eval
import tensorflow.keras.backend as K
import logging

#aggiunte
from sklearn.metrics import accuracy_score, classification_report
#matrice di confusione 

import scikitplot as skplt
import matplotlib.pyplot as plt

# gridsearch
import sklearn
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import classification_report, confusion_matrix 
from sklearn.datasets import load_breast_cancer 
from sklearn.svm import SVC 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split 


h = .02  # step size in the mesh

#Input giulia
raw_covid_142 = pd.read_csv('StatisticalDataofAllPatients_10_90_COVID.csv', sep='\t')
raw_nocovid = pd.read_csv('StatisticalDataofAllPatients_10_90_noCOVID.csv', sep='\t')
raw_covid_57 = pd.read_csv('StatisticalDataofAllPatients_10_90.csv', sep='\t')

#qui sto creando una nuova colonna con 1 quando i pazienti sono covid 
raw_covid_142['covid'] = 1
raw_covid_57['covid'] = 1

#concatena 
raw_covid = pd.concat([raw_covid_142, raw_covid_57])

raw_covid['covid'] = 1
raw_nocovid['covid'] = 0
data = pd.concat([raw_covid, raw_nocovid])

#mischia i dati
data = data.sample(frac=1).reset_index(drop=True)
data['PatientAge'].replace('Y', '', inplace=True, regex=True)

#Converti i dati grezzi in un tipo di dati categorico. 0 per il maschio e 1 per la femmina 
#converte l'et in un interi
data['PatientAge'] = data['PatientAge'].astype(np.int32)
data['PatientSex'] = data['PatientSex'].map({'M': 0, 'F': 1})
tags = ['PatientName', 'PatientID', 'AccessionNumber', 'AcquisitionDate', 'Kernel', 'COVtag', 'p_25th', 'p_50th', 'p_75th', 'p_90th', 'Kurtosis', 'A_f',
       'p_25th_ill', 'p_50th_ill', 'p_75th_ill']




#Rimuovere righe o colonne specificando i nomi delle etichette e l'asse 
#corrispondente oppure specificando direttamente i nomi di indici o colonne.
data = data.drop(tags, axis=1)
data.drop(data[data.Media_ill == 'no Gauss'].index, inplace=True)



#ottieni l'"asse delle informazioni" con keys
for k in data.keys():
    #Ci restituisce una serie con il tipo di dati di ogni 
    #colonna. L'indice del risultato sono le colonne del DataFrame originale. 
    #Le colonne con tipi misti vengono archiviate con objectdtype. 
    
    #se sono colonne miste allora sostituisci con data[k].astype(np.float64)
    if (data[k].dtype == 'object'):
        data[k] = data[k].astype(np.float64)
data
print("numero colonne")
print(len(data.columns))
print ('numero dati dopo modifica: ', data.size/18)

# ## TR-VAL split
# train test
train_dataset = data.sample(frac=0.85, random_state=0)
test_dataset = data.drop(train_dataset.index)
y_train = train_dataset.pop('covid').to_numpy()#train_label
y_test = test_dataset.pop('covid').to_numpy()#test_label
train_stats = train_dataset.describe()
train_stats = train_stats.transpose()
#normalizza
def norm(x):
    return (x - train_stats['mean']) / train_stats['std']

X_train = norm(train_dataset).to_numpy()
X_test= norm(test_dataset).to_numpy()

y_train.shape 
print("y_train")
print(y_train)


h = .02  # step size in the mesh

names1 = ["Nearest Neighbors", "Linear SVM", "RBF SVM", "Gaussian Process",
         "Decision Tree", "Random Forest"]

names2 = ["Neural Net", "AdaBoost",
         "Naive Bayes", "QDA"]
        
classifiers1 = [
    KNeighborsClassifier(3),#controlla i 3 dati vicini
    
    SVC(kernel="linear", C=0.025),#la regolarizzazione  inversamento 
    #proporzionale a C(quando C  grande le barre sono grandi) 
    
    SVC(gamma=2, C=1),#all'aumentare di gamma il sistema prende aree sempre pi strette intorno ai miei dati
    GaussianProcessClassifier(1.0 * RBF(1.0)),
    DecisionTreeClassifier(max_depth=5), # numeri di livelli
    RandomForestClassifier(max_depth=5, n_estimators=10, max_features=1),#n_estimators=10 sono Il numero di alberi 
    #nella foresta.
    #max_features=1 il numero di caratteristiche da considerare quando si cerca la migliore suddivisione
]
classifiers2 = [
    #non cambiare i parametri 
    MLPClassifier(alpha=1, max_iter=1000),#max_iter=1000 Numero massimo di iterazioni. 
    AdaBoostClassifier(),
    GaussianNB(),
    QuadraticDiscriminantAnalysis()
]

Parameters_list1 = [
    
    #KNeighborsClassifier
    {'n_neighbors':np.arange(1,10,1)},#intervallo 1-10
    #SVC(kernel="linear", C=0.025)
    {'kernel':('linear', 'rbf'), 'C':np.linspace(0.02,0.2,101)},#intervallo da 0.02-0.20
    #SVC(gamma=2, C=1)
    {'gamma':np.linspace(0.4,2.0,101),#intervallo tra 0.4-2.0
     'C':np.linspace(0.1,2.0,101)},#intervallo
    #GaussianProcessClassifier(1.0 * RBF(1.0)),
    {'kernel':[1.0 * RBF(1.0),2.0 * RBF(1.0)]},
    #DecisionTreeClassifier(max_depth=5), 
    {'max_depth':np.linspace(1.0,10.0,10)},# numeri di livelli
    #RandomForestClassifier
    {'max_depth':np.arange(1,11,1), 
    'n_estimators':np.arange(1,16,1),
    'max_features':np.arange(1,6,1)}
]
    
# iterate over classifiers
for name, classificatore1, parametri in zip(names1, classifiers1, Parameters_list1): #scorri tutti i classificatori
    clf = GridSearchCV(classificatore1, parametri)
    clf.fit(X_train, y_train)
   
    score_test = clf.score(X_test, y_test) #calcolo accuratezza
    score_train = clf.score(X_train, y_train)
    print (name,clf,"        score test", score_test)
    print (" score train",score_train)
    print('Parametro miglione')
    print(clf.best_params_)
    
    
    #predizione 
    grid_predictionstest = clf.predict(X_test)
    grid_predictionstrain = clf.predict(X_train)
    
    #Confronto le previsioni (grid_predictionstest) con i risultati corretti (y_test)
    report_train=classification_report(grid_predictionstrain, y_train)
    report_test=classification_report(grid_predictionstest, y_test)
    print()
    print('CONFRONTO TRA PREVISIONI E VALORI REALI PER IL TRAIN')
    print(report_train)
    print()
    
    #la matrice di confusione mi permette di capire quali sono le risposte sbagliate del modello.
    skplt.metrics.plot_confusion_matrix(grid_predictionstrain,y_train)
    plt.show()
    print('X_train predetti')
    print(clf.predict(X_train))
    print('y_train')
    print(y_train)
    
    print()
    print('CONFRONTO TRA PREVISIONI VALORI REALI PER IL TEST')
    print(report_test)
    print()
    
    skplt.metrics.plot_confusion_matrix(grid_predictionstest,y_test)
    plt.show()
    print('X_test predetti')
    print(clf.predict(X_test)) 
    print('y_test')
    print(y_test)
    
    
    #codice giulia da aggiungere 

    K.clear_session()
    #p = clf.evaluate(X_train, y_train)[1]
    
    
    # calculate the fpr and tpr for all thresholds of the classification
    #probs = model.predict_proba(X_test)
    
    # test)
    fpr, tpr, threshold = skplt.metrics.roc_curve(y_test,grid_predictionstest )
    roc_auc = skplt.metrics.auc(fpr, tpr)

    # method I: plt
   
    plt.title('Test:Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    #train
    fpr, tpr, threshold = skplt.metrics.roc_curve(y_train,grid_predictionstrain )
    roc_auc = skplt.metrics.auc(fpr, tpr)

    # method I: plt
   
    plt.title('Train:Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print("__________________________________________________________")
    print("__________________________________________________________")
    print("__________________________________________________________")
    print()
    
    
plt.tight_layout()
plt.show()


# iterate over classifiers
for name, clf in zip(names2, classifiers2): #scorri tutti i classificatori
#     clf = GridSearchCV(classificatore2)
    clf.fit(X_train, y_train)
    score_test = clf.score(X_test, y_test) #calcolo accuratezza
    score_train = clf.score(X_train, y_train)
    print (name,": score test", score_test, clf)
    print ("                    score train",score_train)
    #predizione 
    p_test=clf.predict(X_test)
    p_train=clf.predict(X_train)
    #Confronto le previsioni (p_test) con i risultati corretti (y_test)
    report_train=classification_report(p_train, y_train)
    report_test=classification_report(p_test, y_test)
    print()
    print('CONFRONTO TRA PREVISIONI E VALORI REALI PER IL TRAIN')
    print(report_train)
    print()
    
    #la matrice di confusione mi permette di capire quali sono le risposte sbagliate del modello.
    skplt.metrics.plot_confusion_matrix(p_train,y_train)
    plt.show()
    print(clf.predict(X_train)) 
    print('y_train')
    print(y_train)
    
    print()
    print('CONFRONTO TRA PREVISIONI VALORI REALI PER IL TEST')
    print(report_test)
    print()
    
    skplt.metrics.plot_confusion_matrix(p_test,y_test)
    plt.show()
    print(clf.predict(X_test))
    print('y_test')
    print(y_test)
    
    
    #codice giulia da aggiungere 

    K.clear_session()
    #p = clf.evaluate(X_train, y_train)[1]
    
    
    # calculate the fpr and tpr for all thresholds of the classification
    #probs = model.predict_proba(X_test)
    
    # test)
    fpr, tpr, threshold = skplt.metrics.roc_curve(y_test,p_test )
    roc_auc = skplt.metrics.auc(fpr, tpr)

    # method I: plt
   
    plt.title('Test:Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    
    #train
    fpr, tpr, threshold = skplt.metrics.roc_curve(y_train,p_train )
    roc_auc = skplt.metrics.auc(fpr, tpr)

    # method I: plt
   
    plt.title('Train:Receiver Operating Characteristic')
    plt.plot(fpr, tpr, 'b', label = 'AUC = %0.2f' % roc_auc)
    plt.legend(loc = 'lower right')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
    plt.show()
    print("__________________________________________________________")
    print("__________________________________________________________")
    print("__________________________________________________________")
    print()
    
    
plt.tight_layout()
plt.show()

```
